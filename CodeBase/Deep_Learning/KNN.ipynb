```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
x, y= make_classification(n_samples=7000, n_features=2,  random_state=25,n_redundant=0,n_classes=2, class_sep=1.5)
```

```python
import seaborn as sns
plt.figure(figsize=(15,7))
sns.scatterplot(x=x[: ,0], y=x[:, 1], hue=y)
plt.scatter(-1, -3, c='red', s=100)
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.scatter(x[:, 0],x[: ,1],c=y)
plt.show()
```

```python
##creating own KNN algorithm
def custom_KNN(x,y,point,k=int(np.ceil(np.sqrt(x.shape[0])))):
  #step1:calculate the distance of given point from all the point given in dataset
  distance_from_each_point=[] #empty list
  for i in range(x.shape[0]):
    #calculate the distance using euclidean distanc
    #d=sqrt(x2-x1**2 +(y2-y1)**2)
    distance=np.sqrt((point[0]-x[i][0])**2 +(point[1]-x[i][1])**2)
    #append distance into the list
    distance_from_each_point.append([distance, y[i]])

    #step2:sort the list in ascending order
    distance_from_each_point.sort()

    #step3: categorize the points based on distances and k
    distance_from_each_point=distance_from_each_point[:k]

    #step4:identify the maximum occuring category
    count_0,count_1=0,0
    for i in distance_from_each_point:
      if i[1]==0:
        count_0 += 1
      else:
        count_1 += 1

    #step5: Return the category
    if count_0 > count_1:
      return 0
    else:
      return 1

```

```python
custom_KNN(x,y,[-1, -3])
```

```python
#compairing sklearn and custom knn algorithm
from sklearn.neighbors import KNeighborsClassifier
obj= KNeighborsClassifier()
obj.fit(x,y)
obj.predict(np.array([-1, -3]).reshape(1,-1))[0]
```
